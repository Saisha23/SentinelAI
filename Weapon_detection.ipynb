```python
from ultralytics import YOLO

model = YOLO("/content/SentinelYOLO/runs/detect/train/weights/best.pt")
metrics = model.val(data="/content/SentinelYOLO/Weapons-Detection-2-1/data.yaml")

print("Precision =", metrics.box.p)
print("Recall    =", metrics.box.r)
print("mAP50     =", metrics.box.map50)
print("mAP50-95  =", metrics.box.map)
```

```python
from ultralytics import YOLO

model = YOLO("/content/SentinelYOLO/runs/detect/train/weights/best.pt")
train_metrics = model.val(split='train')

print("==== TRAIN METRICS ====")
print("Precision  =", train_metrics.results_dict['metrics/precision(B)'])
print("Recall     =", train_metrics.results_dict['metrics/recall(B)'])
print("mAP50      =", train_metrics.results_dict['metrics/mAP50(B)'])
print("mAP50-95   =", train_metrics.results_dict['metrics/mAP50-95(B)'])
print("========================")
```

```python
model = YOLO("/content/SentinelYOLO/runs/detect/train/weights/best.pt")
val_metrics = model.val(split='val')

print("==== VALIDATION METRICS ====")
print("Precision  =", val_metrics.results_dict['metrics/precision(B)'])
print("Recall     =", val_metrics.results_dict['metrics/recall(B)'])
print("mAP50      =", val_metrics.results_dict['metrics/mAP50(B)'])
print("mAP50-95   =", val_metrics.results_dict['metrics/mAP50-95(B)'])
print("============================")
```

```python
model = YOLO("/content/SentinelYOLO/runs/detect/train/weights/best.pt")
test_metrics = model.val(split='test')

print("==== TEST METRICS ====")
print("Precision  =", test_metrics.results_dict['metrics/precision(B)'])
print("Recall     =", test_metrics.results_dict['metrics/recall(B)'])
print("mAP50      =", test_metrics.results_dict['metrics/mAP50(B)'])
print("mAP50-95   =", test_metrics.results_dict['metrics/mAP50-95(B)'])
print("=======================")
```
